```python
sales = pd.read_csv("historic_sales.csv",delimiter='|',index_col=0,low_memory=False)

sales['Dept'] = pd.to_numeric(sales['Dept'], errors='coerce').fillna(0).astype('int')
# sales.dtypes

sales['Weekly_Sales'] = sales['Weekly_Sales'].str.replace(',','.')
sales['Weekly_Sales'] = pd.to_numeric(sales['Weekly_Sales'], errors='coerce')
# sales['Weekly_Sales']

sales['IsHoliday'] = sales['IsHoliday'].astype('bool')

# sales.dtypes
sales
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

# Code from https://seaborn.pydata.org/examples/many_pairwise_correlations.html
sns.set(style="white")

corr = merged_sales.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
